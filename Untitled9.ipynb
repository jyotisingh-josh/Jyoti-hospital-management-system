```python
def add_patient():
    patient_id = input("Enter Patient ID: ")
    patient_name = input("Enter Patient Name: ")
    phone = input("Enter Phone Number: ")
    disease = input("Enter Disease: ")

    file = open("patients.txt","a")
    file.write(f"{patient_id},{patient_name},{phone},{disease}\n")
    file.close()

    print("Patient Added Successfully")
def display_patients():
    print("PATIENT LIST")
    print("----------------")

    file = open("patients.txt","r")
    print(file.read())
    file.close()
def update_patient():
    pid = input("Enter Patient ID to update: ")
    file = open("patients.txt","r")
    lines = file.readlines()
    file.close()

    file = open("patients.txt","w")

    for line in lines:
        data = line.strip().split(",")

        if data[0] == pid:
            name = input("Enter new name: ")
            phone = input("Enter new phone: ")
            disease = input("Enter new disease: ")

            file.write(f"{pid},{name},{phone},{disease}\n")
        else:
            file.write(line)

    file.close()
    print("Patient Updated")
def delete_patient():
    pid = input("Enter Patient ID to delete: ")

    file = open("patients.txt","r")
    lines = file.readlines()
    file.close()

    file = open("patients.txt","w")

    for line in lines:
        if not line.startswith(pid):
            file.write(line)

    file.close()
    print("Patient Deleted")
def add_doctor():
    doc_id = input("Enter Doctor ID: ")
    doc_name = input("Enter Doctor Name: ")
    specialization = input("Enter Specialization: ")

    file = open("doctors.txt","a")
    file.write(f"{doc_id},{doc_name},{specialization}\n")
    file.close()

    print("Doctor Added Successfully")
def display_doctors():

    print("DOCTORS")
    print("----------------")

    file = open("doctors.txt","r")
    print(file.read())
    file.close()
def book_appointment():

    import random

    appointment_id = random.randint(100,999)
    patient_id = input("Enter Patient ID: ")
    doctor_id = input("Enter Doctor ID: ")
    date = input("Enter Appointment Date: ")

    file = open("appointments.txt","a")
    file.write(f"{appointment_id},{patient_id},{doctor_id},{date}\n")
    file.close()

    print("Appointment Booked Successfully")
def display_appointments():

    print("APPOINTMENTS")
    print("----------------")

    file = open("appointments.txt","r")
    print(file.read())
    file.close()
def add_payment():

    import random

    payment_id = random.randint(1000,9999)
    patient_id = input("Enter Patient ID: ")
    amount = input("Enter Amount: ")
    method = input("Enter Payment Method: ")

    file = open("payments.txt","a")
    file.write(f"{payment_id},{patient_id},{amount},{method}\n")
    file.close()

    print("Payment Successful")
def display_payments():

    print("PAYMENTS")
    print("----------------")

    file = open("payments.txt","r")
    print(file.read())
    file.close()
def support_request():

    import random

    ticket_id = random.randint(1,500)
    patient_id = input("Enter Patient ID: ")
    issue = input("Enter your issue: ")

    file = open("support.txt","a")
    file.write(f"{ticket_id},{patient_id},{issue},OPEN\n")
    file.close()

    print("Support Ticket Created")
def patient_menu():

    while True:

        print("\nPATIENT MENU")
        print("1 View Patients")
        print("2 Book Appointment")
        print("3 Make Payment")
        print("4 Request Support")
        print("5 Exit")

        choice = int(input("Enter choice: "))

        if choice == 1:
            display_patients()

        elif choice == 2:
            book_appointment()

        elif choice == 3:
            add_payment()

        elif choice == 4:
            support_request()

        else:
            break
def admin_menu():

    while True:

        print("\nHOSPITAL MANAGEMENT MENU")

        print("1 Add Patient")
        print("2 Display Patients")
        print("3 Update Patient")
        print("4 Delete Patient")

        print("5 Add Doctor")
        print("6 Display Doctors")

        print("7 Book Appointment")
        print("8 Display Appointments")

        print("9 Display Payments")

        print("10 Exit")

        choice = int(input("Enter choice: "))

        if choice == 1:
            add_patient()

        elif choice == 2:
            display_patients()

        elif choice == 3:
            update_patient()

        elif choice == 4:
            delete_patient()

        elif choice == 5:
            add_doctor()

        elif choice == 6:
            display_doctors()

        elif choice == 7:
            book_appointment()

        elif choice == 8:
            display_appointments()

        elif choice == 9:
            display_payments()

        else:
            break
def run_system():

    print("1 Hospital Admin")
    print("2 Patient")

    choice = int(input("Enter choice: "))

    if choice == 1:
        admin_menu()

    elif choice == 2:
        patient_menu()

    else:
        print("Invalid Choice")
run_system()
```
